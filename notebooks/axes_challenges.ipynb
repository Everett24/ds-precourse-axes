````python
# Your code below (You need to make the appropriate imports of matplotlib and numpy)
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

earnings_data = '''
0.71  0.63  0.85  0.44
0.61  0.69  0.92  0.55
0.72  0.77  0.92  0.60
0.83  0.80  1.00  0.77
0.92  1.00  1.24  1.00
1.16  1.30  1.45  1.25
1.26  1.38  1.86  1.56
1.53  1.59  1.83  1.86
1.53  2.07  2.34  2.25
2.16  2.43  2.70  2.25
2.79  3.42  3.69  3.60
3.60  4.32  4.32  4.05
4.86  5.04  5.04  4.41
5.58  5.85  6.57  5.31
6.03  6.39  6.93  5.85
6.93  7.74  7.83  6.12
7.74  8.91  8.28  6.84
9.54 10.26  9.54  8.73
11.88 12.06 12.15  8.91
14.04 12.96 14.85  9.99
16.20 14.67 16.02 11.61
'''
earnings_data_str = earnings_data.replace('  ',' ')
earnings_arr = np.array(earnings_data_str.split())
earnings_arr = earnings_arr.astype(float)
quarters = np.array([np.linspace(1960,1980.75,num=84)])
qt = np.transpose(quarters)
plt.plot(qt, earnings_arr, '-')
#plt.xticks(np.arange(min(x), max(x)+1, 2.0))
plt.xticks(rotation=45)
plt.title("Johnson and Johnson Quarterly Earnings")
plt.ylabel("Earnings per share (U.S. Dollars)")
plt.xlabel("Year")
plt.show()
````

1. Modify the given plot to set the limits of the x-axis at `(-4, 4)`, and the y-axis to `(-8, 8)`
2. Modify the given plot to have ticks and labels at every odd number, and zero of each axis.

````python
# Modify this code to alter the axes

import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(0, 2, 1000)
y = np.random.normal(0, 4, 1000)
plt.plot(x, y, 'ok', markersize=3, alpha=0.2)
plt.xlim(-4,4)
plt.ylim(-8,8)

#adjust ticks
xodds = [-3, -1, 0, 1, 3]
yodds = [-7, -5, -3, -1, 0, 1, 3, 5, 7]
plt.xticks(xodds)
plt.yticks(yodds)

plt.show()
````
